### Model

```python
DATADIR = './data/gdb13/'
OUTDIR = './runs/run11/'
LOGFILE = os.path.join(OUTDIR, 'log.csv')
DATASPEC_FILE = os.path.join(DATADIR, 'spec.json')
RUNSPEC_FILE = os.path.join(OUTDIR, 'runspec.json')

dataspec = fetch_params(DATASPEC_FILE)
if os.path.isfile(RUNSPEC_FILE):
    runspec = fetch_params(RUNSPEC_FILE)
    L = runspec.L
else:
    L = 128

encoder = Encoder(L)
encoder.load_state_dict(torch.load(OUTDIR + 'encoder_weights.pth', map_location=torch.device('cpu')))
encoder.eval()

decoder = Decoder(L)
decoder.load_state_dict(torch.load(OUTDIR + 'decoder_weights.pth', map_location=torch.device('cpu')))
decoder.eval()

predictor = Predictor(L)
predictor.load_state_dict(torch.load(OUTDIR + 'predictor_weights.pth', map_location=torch.device('cpu')))
predictor.eval()
None
```

### Data

```python
trainset, testset = make_data('./data/gdb13', n=1000)

means, stds = get_latent_distributions(encoder, testset.hots)

compressed_dims = get_important_dimensions(testset.hots, encoder, decoder)

compressed_means = means[compressed_dims]
compressed_stds = stds[compressed_dims]

bounds = torch.stack((
    (compressed_means - 4 * compressed_stds), 
    (compressed_means + 4 * compressed_stds)
))

unit_bounds = torch.zeros_like(bounds)
unit_bounds[1] = 1

plt.scatter(testset.qed, -testset.sas, s=1)
plt.xlabel('qed')
plt.ylabel('-sas')
plt.xlim([0, 1])
plt.ylim([-10, 0])
```

### Objective Score Function

```python
def filter_and_score(candidates, compressed=True):
    if compressed: candidates = decompress(candidates, compressed_dims, L)
    
    smiles = from_hot(decoder(candidates), dataspec.alphabet)
    
    x = []
    scores = []
    for i, smile in enumerate(smiles):
        mol = MolFromSmiles(smile)
        if mol is None: continue
        
        x.append(candidates[i, compressed_dims].unsqueeze(0))
        scores.append([QED(mol), -SAS(mol)])
        
    if len(x) > 0:
        return torch.cat(x), torch.Tensor(scores)
    else:
        return torch.Tensor(), torch.Tensor()
```

### Sample

```python
def get_samples(n=16, qed_lim=(0, 0.5), sas_lim=(5, 10)):
    
    z = torch.Tensor()
    scores = torch.Tensor()
    
    indices = list(range(len(testset)))
    random.shuffle(indices)
    
    for i in indices:
        
        if z.size(0) >= n: break
        
        qed = testset.qed[i]
        sas = testset.sas[i]
        
        if qed > qed_lim[0] and qed < qed_lim[1] and sas > sas_lim[0] and sas < sas_lim[1]:
            sample_z, _, _ = encoder(testset.hots[i:i+1])
            z = torch.cat([z, sample_z])
            scores = torch.cat(
                [scores, 
                 torch.cat([testset.qed[i], -testset.sas[i]]).unsqueeze(0)
                ]
            )

    return z[:, compressed_dims].detach(), scores.detach()
```

### Surrogate

```python
def fit_surrogate(x, scores, state_dict):
    x = normalize(x, bounds)
    models = [
        SingleTaskGP(
            train_X=x.detach(), 
            train_Y=scores[..., i:i+1].detach(), 
            outcome_transform=Standardize(m=1)
        ) for i in range(scores.size(-1))
    ]
    
    model = ModelListGP(*models)
    
    if state_dict is not None:
        model.load_state_dict(state_dict)

    mll = SumMarginalLogLikelihood(model.likelihood, model)
    
    fit_gpytorch_mll(mll)
    
    return model
```

### Acquisition Function

```python
def optimize_qehvi(model, x, scores, ref_point):
    
    x = normalize(x, bounds)
    with torch.no_grad():
        pred = model.posterior(x).mean

    partitioning = FastNondominatedPartitioning(
        ref_point=ref_point,
        Y=pred,
    )
    
    sampler = SobolQMCNormalSampler(sample_shape=torch.Size([64]))
    
    acq_func = qExpectedHypervolumeImprovement(
        model=model,
        ref_point=ref_point,
        partitioning=partitioning,
        sampler=sampler,
    )

    candidates, _ = optimize_acqf(
        acq_function=acq_func,
        bounds=unit_bounds,
        q=8,
        num_restarts=8, 
        raw_samples=64 
    )
    candidates = unnormalize(candidates.detach(), bounds=bounds)
    
    return filter_and_score(candidates)

def optimize_qnparego(model, x, scores):
    
    x = normalize(x, bounds)
    with torch.no_grad():
        pred = model.posterior(x).mean
    
    sampler = SobolQMCNormalSampler(sample_shape=torch.Size([128]))
    
    acq_func_list = []

    for _ in range(8):
        weights = sample_simplex(scores.size(-1)).squeeze()
        
        objective = GenericMCObjective(
            get_chebyshev_scalarization(weights=weights, Y=pred)
        )
        
        acq_func = qNoisyExpectedImprovement(
            model=model,
            objective=objective,
            X_baseline=x,
            sampler=sampler,
            prune_baseline=True,
        )
        
        acq_func_list.append(acq_func)
    
    # optimize
    candidates, _ = optimize_acqf_list(
        acq_function_list=acq_func_list,
        bounds=unit_bounds,
        q=8, 
        num_restarts=8, 
        raw_samples=64,  # used for intialization heuristic
    )
    candidates = unnormalize(candidates.detach(), bounds=bounds)
    
    return filter_and_score(candidates)
```

### Optimization Loop

```python
# Load saved progress
if True:
    candidates_x = torch.load('./bo/candidates_x.pt')
    candidates_scores = torch.load('./bo/candidates_scores.pt')
    
    samples_x = torch.load('./bo/samples_x.pt')
    samples_scores = torch.load('./bo/samples_scores.pt')
    
    state_dict = torch.load('./bo/state_dict.pt')

# Start New Run
else:
    candidates_x, candidates_scores = torch.Tensor(), torch.Tensor()
    samples_x, samples_scores = torch.Tensor(), torch.Tensor()
    state_dict = None
    
hypervolumes = []
```

```python
%%capture --no-display

qed_lim = (0, 1)
sas_lim = (0, 10)

# Samples From Dataset
sample_x, sample_scores = get_samples(n=16, qed_lim=qed_lim, sas_lim=sas_lim)
samples_x = torch.cat([samples_x, sample_x])
samples_scores = torch.cat([samples_scores, sample_scores])

# Candidate Samples Acquired During Optimization
candidates_x = torch.cat([candidates_x, sample_x])
candidates_scores = torch.cat([candidates_scores, sample_scores])

# Baseline For Optimization
ref_point = samples_scores.min(0)[0]

for iteration in range(16):
    
    x = torch.cat([samples_x, candidates_x])
    scores = torch.cat([samples_scores, candidates_scores])
    
    # Fit GP surrogate
    surrogate = fit_surrogate(x, scores, state_dict)
    state_dict = surrogate.state_dict()
    
    # Acquire candidate samples
    new_x, new_scores = optimize_qehvi(surrogate, x, scores, ref_point)
    
    # Update samples
    candidates_x = torch.cat([candidates_x, new_x])
    candidates_scores = torch.cat([candidates_scores, new_scores])
    
    # Update hypervolume Progress
    bd = DominatedPartitioning(ref_point=ref_point, Y=scores)
    hypervolumes.append(bd.compute_hypervolume().item())
    
plt.plot(hypervolumes)
```
